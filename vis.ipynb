```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
plt.rcParams['font.sans-serif'] = 'Times New Roman'

file_names = [ './wbp-result.mat','./wrnn-result.mat', './wgru-result.mat', './wtransformer.mat','./winformer.mat', './wautoformer.mat', 
             './wfedformer.mat','./wdlinear.mat', './wlstm-result.mat','./wbsa-result.mat', './wstlga-result.mat',
   './wstlpso-result.mat','./wzoa.mat','./wooa.mat','./wstlbsa-result.mat','./waln-result.mat', './wstlaln-result.mat']

target = []
preds = []
errors = []
for idx in range(17):
    true = loadmat(file_names[idx])['true']
    target.append(np.array(true).reshape(-1))
    pred = loadmat(file_names[idx])['pred']
    preds.append(np.array(pred).reshape(-1))
    errors.append(np.array(true).reshape(-1) - np.array(pred).reshape(-1))
```
